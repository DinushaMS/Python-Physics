```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
from functions import calcPhonon
```

```python
#Phonon dispersion of an fcc crystal
#http://lampx.tugraz.at/~hadley/ss1/phonons/fcc/fcc.php

arrX = []
arrW1 = []
arrW2= []
arrW3 = []

arrWS = [[0,0,0],[2*np.pi,0,0],[2*np.pi,np.pi,0],[3*np.pi/2,3*np.pi/2,0],[0,0,0],[np.pi,np.pi,np.pi]]
cn=200
n = 0
for sec in range(len(arrWS)-1):
    x = np.linspace(arrWS[sec][0],arrWS[sec+1][0],cn)
    y = np.linspace(arrWS[sec][1],arrWS[sec+1][1],cn)
    z = np.linspace(arrWS[sec][2],arrWS[sec+1][2],cn)
    for cnt in range(cn):
        i,j,k=x[cnt],y[cnt],z[cnt]
        x1,x2,x3=calcPhonon(i,j,k)
        arrX.append(n)
        arrW1.append(x1)
        arrW2.append(x2)
        arrW3.append(x3)
        #print()
        n+=1

plt.figure(figsize=(10,6))
plt.plot(arrX,arrW1)
plt.plot(arrX,arrW2)
plt.plot(arrX,arrW3)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(np.concatenate((arrW1, arrW2, arrW3), axis=None),bins=100,histtype='bar')
plt.show()
```
