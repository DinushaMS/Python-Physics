simple harmonic oscillator
$$\frac{d^2\theta}{dt^2}+\omega \theta = 0$$
$$\frac{d\theta}{dt}=y$$
$$\frac{dy}{dt}=-\omega \theta$$

```python
omega = 1.0  # Angular frequency
#args = [x,y]
def dxdt(t, *args):
    return args[1]

def dydt(t, *args):
    #angula frequency
    return -omega*args[0]

t = np.linspace(0, 5, 1000)
y = np.zeros(len(t))
x = np.zeros(len(t))
a = np.zeros(len(t))

h = t[1] - t[0]
x[0] = 0.0  # Initial condition for x, angle
y[0] = 1.0  # Initial condition for y, veleocity
a[0] = -omega*x[0]  # Initial condition for acceleration

for i in range(1, len(t)):
    x[i], y[i] = RK4_step([dxdt, dydt], t[i-1], h, x[i-1], y[i-1])
    a[i] = -omega*x[i]


plt.plot(t, x, label='x(t)')
plt.plot(t, y, label='y(t)')
plt.plot(t, a, label='a(t)')
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Runge-Kutta 4th Order Method')
plt.legend()
plt.grid()
plt.show()
```
