```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
data = np.loadtxt('Thorlabs_NDF_specs.csv', delimiter=',', skiprows=1)
wavelengths = data[:, 0]
correct_att_10 = data[:, 1]
correct_att_30 = data[:, 2]
correct_att_100 = data[:, 3]
correct_att_1000 = data[:, 4]

mono_wl = 738
# scipy interpolation of the attenuation values at the monochromator wavelength
mono_att_1000 = interpolate.interp1d(wavelengths, correct_att_10)(mono_wl)
mono_att_100 = interpolate.interp1d(wavelengths, correct_att_30)(mono_wl)
mono_att_30 = interpolate.interp1d(wavelengths, correct_att_100)(mono_wl)
mono_att_10 = interpolate.interp1d(wavelengths, correct_att_1000)(mono_wl)
correct_att = [mono_att_10, mono_att_30, mono_att_100, mono_att_1000]
```

```python
# set of attenuations used
att_1 = [10,30,100,1000]
#att_1 = [10, 30, 120, 650]
selected_total_attenuation = []
# select 3 attenuators from att_1 and appnd to a list
selected_attenuators = []
for att in att_1:
    selected_attenuators.append([att, 1, 1])
    selected_total_attenuation.append(att)
for combination in itertools.combinations(att_1, 2):
    selected_attenuators.append([combination[0], combination[1], 1])
    selected_total_attenuation.append(combination[0] * combination[1])
for combination in itertools.combinations(att_1, 3):
    selected_attenuators.append([combination[0], combination[1], combination[2]])
    selected_total_attenuation.append(combination[0] * combination[1] * combination[2])
```

```python
target_att = 1000
# find indexs of the selected_total_attenuation that are equal to the target_att
indexs = [i for i, x in enumerate(selected_total_attenuation) if x == target_att]
# print the selected attenuators that correspond to the indexs
print("Selected attenuators that correspond to the target attenuation of {}:".format(target_att))
for index in indexs:
    sel_cor_att = []
    for att in selected_attenuators[index]:
            # find the index of the att in att_1
            if att != 1:
                att_index = att_1.index(att)
                sel_cor_att.append(float(correct_att[att_index]))
            else:
                sel_cor_att.append(1)
    print(selected_attenuators[index])
    print(sel_cor_att)
```
