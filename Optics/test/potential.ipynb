```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.arange(-60,300,0.1)
Vh = (0.02*r)**2
Vah = (1-np.exp(-0.02*(r)))**2

plt.figure(figsize=(15,3))
plt.subplot(131)
plt.plot(r,Vah, label='Morse potential')
plt.plot(r,Vh, label='Harmonic potential')
plt.ylim([0,1.2])
plt.legend()
plt.title(r'$V$')
plt.subplot(132)
plt.plot(r[:-1], np.diff(Vah,1), label='Morse potential')
plt.plot(r[:-1], np.diff(Vh,1), label='Harmonic potential')
plt.legend()
plt.title(r'$\frac{dV}{dr}$')
plt.subplot(133)
plt.plot(r[:-2], np.diff(Vah,2), label='Morse potential')
plt.plot(r[:-2], np.diff(Vh,2), label='Harmonic potential')
plt.legend()
plt.title(r'$\frac{d^2V}{dr^2}$')
```
