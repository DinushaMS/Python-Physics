```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
folder_path = "C:\\Users\\dinus\\Downloads\\"
file_names = ["CARS_spectrum_td=2fs.dat", "CARS_spectrum_td=302fs.dat", "CARS_spectrum_td=1002fs.dat", "CARS_spectrum_td=1502fs.dat", "CARS_spectrum_td=2002fs.dat"]
mult_factors = [1, 1,1,1,1]
time_delays = [0, 300, 1000, 1500, 2000]  # in fs

for file_name in file_names:
    file_path = os.path.join(folder_path, file_name)
    data = np.loadtxt(file_path, delimiter=',')
    time = data[:, 0]
    intensity = (data[:, 1]-np.min(data[:, 1]))*mult_factors[file_names.index(file_name)]
    plt.plot(time, intensity, label= f"x{mult_factors[file_names.index(file_name)]}, "+str(time_delays[file_names.index(file_name)])+" fs")
plt.xlabel('Time (fs)')
plt.ylabel('Intensity (a.u.)')
plt.legend()
plt.show()
```
