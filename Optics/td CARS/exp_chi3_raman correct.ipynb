# Corrections

$$E_i=\frac{1}{2}E_{i0}e^{i(k_iz-\omega_it)}+c.c.$$

$$I_i=\frac{1}{2}n_i\epsilon_0c|E_{i0}|^2$$

```python
import sympy as sp
```

```python
x,y,z,t = sp.symbols('x y z t', real = True)
E10, E20, E30, Eas0 = sp.symbols('E_{10} E_{20} E_{30} E_{as0}')
Eas0 = sp.Function('E_{as0}')(z,t)
k1, k2, k3, kas, delk, om1, om2, om3, omas, wlas = sp.symbols('k_1 k_2 k_3 k_{as} \\Delta{k} \\omega_1 \\omega_2 \\omega_3 \\omega_{as} \\lambda_{as}',real = True, nonzero=True)
eps0, chi3 = sp.symbols('\\epsilon_0 \\chi^{(3)}', real = True)
n1, n2, n3, nas, c, L = sp.symbols('n_1 n_2 n_3 n_{as} c L', real = True, nonzero=True, positive = True)
xi1,xi2,xi3,xias,tau,w0 = sp.symbols('xi_1 xi_2 xi_3 xi_as tau w_0', real=True, positive=True)
A1, A2, A3, Aas = sp.symbols('A_1 A_2 A_3 A_{as}', real=True, positive=True)
I1, I2, I3, Ias = sp.symbols('I_1 I_2 I_3 I_{as}', real=True, positive=True)
E1, E2, E3, Eas = sp.symbols('E_1 E_2 E_3 E_{as}', real=True, positive=True)
```

```python
E1_expr = sp.Rational(1,2)*E10*sp.exp(sp.I*(k1*z-om1*t))
E1_expr = E1_expr + sp.conjugate(E1_expr)
E2_expr = sp.Rational(1,2)*E20*sp.exp(sp.I*(k2*z-om2*t))
E2_expr = E2_expr + sp.conjugate(E2_expr)
E3_expr = sp.Rational(1,2)*E30*sp.exp(sp.I*(k3*z-om3*t))
E3_expr = E3_expr + sp.conjugate(E3_expr)
Eas_expr = sp.Rational(1,2)*Eas0*sp.exp(sp.I*(kas*z-omas*t))
Eas_expr = Eas_expr + sp.conjugate(Eas_expr)
sp.Eq(Eas,Eas_expr)
```

```python
E_tot_expr = E1_expr+E2_expr+E3_expr
P3_tot_expr = (eps0*chi3*E_tot_expr**3).expand()
sel_E3_product_expr = E10*sp.conjugate(E20)*E30
P3_as_expr = P3_tot_expr.subs(sel_E3_product_expr,1).subs({E10:0, E20:0, E30:0})*sel_E3_product_expr
P3_as_expr = P3_as_expr.subs(sp.exp(-sp.I*(om1-om2+om3)*t).expand(),sp.exp(-sp.I*omas*t))
sp.Eq(sp.symbols('P^{(3)}_{as}(t)'),P3_as_expr)
```

```python
ddEasdz_expr = sp.diff(Eas_expr,z,2)
sp.Eq(sp.symbols('\\nabla^2E_{as}'),ddEasdz_expr)
```

```python
ddEasdz_re_expr = ddEasdz_expr.subs({sp.conjugate(Eas0):0, sp.diff(sp.conjugate(Eas0),z,1):0, sp.diff(sp.conjugate(Eas0),z,2):0}).collect(sp.exp(sp.I*(omas*t-kas*z)))
ddEasdz_re_expr = ddEasdz_re_expr.subs(sp.diff(Eas0,z,2),0)
sp.Eq(sp.symbols('\\nabla^2E_{as}'),ddEasdz_re_expr)
```

```python
ddEasdt_expr = sp.diff(Eas_expr,t,2)
sp.Eq(sp.symbols('\\frac{\\partial^2E_{as}}{\\partial{t^2}}'),ddEasdt_expr)
```

```python
ddEasdt_re_expr = ddEasdt_expr.subs({sp.conjugate(Eas0):0, sp.diff(sp.conjugate(Eas0),t,1):0, sp.diff(sp.conjugate(Eas0),t,2):0}).collect(sp.exp(sp.I*(omas*t-kas*z)))
ddEasdt_re_expr = ddEasdt_re_expr.subs(sp.diff(Eas0,t,2),0)
sp.Eq(sp.symbols('\\frac{\\partial^2E_{as}}{\\partial{t^2}}'),ddEasdt_re_expr)
```

```python
ddP3_asdt_expr = sp.diff(P3_as_expr,t,2).simplify()
sp.Eq(sp.symbols('\\frac{\\partial^2P^{(3)}_{as}}{\\partial{t^2}}'),ddP3_asdt_expr)
```

```python
#wave equation
we_lhs_expr = (ddEasdz_re_expr-nas**2*ddEasdt_re_expr/c**2)
we_rhs_expr = ddP3_asdt_expr/eps0/c**2
sp.Eq(we_lhs_expr,we_rhs_expr)
```

```python
we_lhs_expr = (we_lhs_expr*sp.exp(-sp.I*(kas*z-omas*t))*2).expand()
we_rhs_expr = (we_rhs_expr*sp.exp(-sp.I*(kas*z-omas*t))*2).expand().simplify()
we_rhs_expr = we_rhs_expr.subs(k1-k2+k3-kas,delk ,nonzero=True)
sp.Eq(we_lhs_expr,we_rhs_expr)
```

```python
#time evolution of Eas is smaller than spatial evolution
we_lhs_expr = we_lhs_expr.subs(sp.diff(Eas0,t,1),0)*(-sp.I/2/kas)
we_rhs_expr = we_rhs_expr*(-sp.I/2/kas)
sp.Eq(we_lhs_expr,we_rhs_expr)
```

```python
#integrate wrt z from 0 to L
we_lhs_int_expr = sp.Integral(we_lhs_expr,(z,0,L)).doit()
we_rhs_int_expr = sp.Integral(we_rhs_expr,(z,0,L)).doit().simplify()
sp.Eq(we_lhs_int_expr,we_rhs_int_expr)
```

```python
#let Eas(0,t) = 0
we_lhs_int_expr = we_lhs_int_expr.subs(Eas0.subs(z,0),0)
sp.Eq(we_lhs_int_expr,we_rhs_int_expr)
```

```python
#intensity############################################################
Ias_expr = sp.Rational(1,2)*nas*eps0*c*we_rhs_int_expr*sp.conjugate(we_rhs_int_expr)
#intensity############################################################
Ias_expr = Ias_expr.subs(E10*sp.conjugate(E10),sp.Abs(E10)**2)
Ias_expr = Ias_expr.subs(E20*sp.conjugate(E20),sp.Abs(E20)**2)
Ias_expr = Ias_expr.subs(E30*sp.conjugate(E30),sp.Abs(E30)**2)
sp.Eq(sp.symbols('I_{as}'),Ias_expr)
```

```python
temp = sp.exp(sp.I*delk*L)-1
temp = (temp*sp.conjugate(temp)).expand()
temp = temp.subs(sp.exp(sp.I*delk*L),sp.cos(delk*L)+sp.I*sp.sin(delk*L)).simplify()
sp.Eq(sp.Abs(sp.exp(sp.I*delk*L)-1)**2,temp)
```

```python
Ias_expr2 = Ias_expr.subs(sp.exp(sp.I*delk*L),sp.cos(delk*L)+sp.I*sp.sin(delk*L)).simplify()
sp.Eq(Ias,Ias_expr2)
```

```python
###################REWRITE ABOVE RESULT##########################
Ias_expr3 = 9*chi3**2*eps0*omas**4*nas*sp.Abs(E10**2*E20**2*E30**2)*sp.sin(L*delk/2)**2/(8*delk**2*c**3*kas**2)
sp.Eq(sp.symbols('I_{as}'),Ias_expr3)
```

```python
Ias_in_E_expr = Ias_expr3.subs(sp.sin(L*delk/2)**2,(L*delk/2)**2*sp.sinc(L*delk/2)**2)
sp.Eq(sp.symbols('I_{as}'),Ias_in_E_expr)
```

```python
E10_expr = A1*sp.exp(-2*sp.log(2)*t**2/tau**2)*sp.exp(-(x**2+y**2)/w0**2)
E20_expr = A2*sp.exp(-2*sp.log(2)*t**2/tau**2)*sp.exp(-(x**2+y**2)/w0**2)
E30_expr = A3*sp.exp(-2*sp.log(2)*t**2/tau**2)*sp.exp(-(x**2+y**2)/w0**2)
sp.Eq(E10, E10_expr)
```

```python
Ias_in_A_expr = Ias_in_E_expr.subs({E10:E10_expr,E20:E20_expr,E30:E30_expr})
sp.Eq(sp.symbols('I_{as}'),Ias_in_A_expr)
```

```python
sp.Eq(xias, sp.Integral(sp.Integral(sp.Integral(Ias,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo)))
```

```python
int1 = sp.Integral(sp.Integral(sp.Integral(sp.exp(-12*sp.log(2)*t**2/tau**2)*sp.exp(-6*(x**2+y**2)/w0**2),(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
sp.Eq(int1,int1.doit())
```

```python
xias_expr = sp.Integral(sp.Integral(sp.Integral(Ias_in_A_expr,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
sp.Eq(xias, xias_expr)
```

```python
xias_expr = xias_expr.doit()
sp.Eq(xias, xias_expr)
```

```python
#intensity############################################################
I1_in_E_expr = sp.Rational(1,2)*n1*eps0*c*sp.Abs(E10)**2
I2_in_E_expr = sp.Rational(1,2)*n2*eps0*c*sp.Abs(E20)**2
I3_in_E_expr = sp.Rational(1,2)*n3*eps0*c*sp.Abs(E30)**2
#intensity############################################################
sp.Eq(I1, I1_in_E_expr)
```

```python
E10_abs_sq_in_I_expr = I1/I1_in_E_expr*sp.Abs(E10)**2
E20_abs_sq_in_I_expr = I2/I2_in_E_expr*sp.Abs(E20)**2
E30_abs_sq_in_I_expr = I3/I3_in_E_expr*sp.Abs(E30)**2
sp.Eq(sp.Abs(E10)**2, E10_abs_sq_in_I_expr)
```

```python
Ias_in_I_expr = sp.simplify(Ias_in_E_expr/sp.Abs(E10)**2/sp.Abs(E20)**2/sp.Abs(E30)**2)*E10_abs_sq_in_I_expr*E20_abs_sq_in_I_expr*E30_abs_sq_in_I_expr
sp.Eq(Ias,Ias_in_I_expr)
```

```python
I1_in_A_expr = I1_in_E_expr.subs(E10,E10_expr)
I2_in_A_expr = I2_in_E_expr.subs(E20,E20_expr)
I3_in_A_expr = I3_in_E_expr.subs(E30,E30_expr)
sp.Eq(I1, I1_in_A_expr)
```

```python
xi1_expr = sp.Integral(sp.Integral(sp.Integral(I1_in_A_expr,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
xi2_expr = sp.Integral(sp.Integral(sp.Integral(I2_in_A_expr,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
xi3_expr = sp.Integral(sp.Integral(sp.Integral(I3_in_A_expr,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
sp.Eq(xi1, xi1_expr)
```

```python
xi1_expr = xi1_expr.doit()
xi2_expr = xi2_expr.doit()
xi3_expr = xi3_expr.doit()
sp.Eq(xi1, xi1_expr)
```

```python
xias_expr_f = xias_expr*(xi1/xi1_expr)*(xi2/xi2_expr)*(xi3/xi3_expr)
sp.Eq(xias, xias_expr_f)
```

```python
xias_expr_ff = xias_expr_f.subs(kas,nas*omas/c)
sp.Eq(xias, xias_expr_ff)
```

```python
xias_expr_fff = xias_expr_ff.subs(omas,2*sp.pi*c/wlas)
sp.Eq(xias, xias_expr_fff)
```

```python
tau_val = 100e-15  # seconds
w0_val = (0.35e-6)/2  # meters
chi3_val = 1e-23  # m^2/V^2
L_val = 0.7e-6  # meters
wlas_val = 700e-9  # meters
n1_val, n2_val, n3_val, nas_val = 2.5, 2.5, 2.5, 2.5  # refractive indices
xi1_val, xi2_val, xi3_val = 1e-9, 1e-9, 1e-9  # pulse energy in J
delk_val = 0  # phase mismatch
eps0_val = 8.854187817e-12  # F/m
c_val = 3e8  # m/s

xias_num = xias_expr_fff.evalf(subs={tau:tau_val, w0:w0_val, chi3:chi3_val, L:L_val, wlas:wlas_val, n1:n1_val, n2:n2_val, n3:n3_val, nas:nas_val, xi1:xi1_val, xi2:xi2_val, xi3:xi3_val, delk:delk_val, eps0:eps0_val, c:c_val})
sp.Eq(xias, xias_num)
```

```python
h_bar = 1.0545718e-34  # JÂ·s
omega_as_val = 2*sp.pi*c_val/wlas_val  # rad/s
N_photons_per_pulse = (xias_num/(h_bar*omega_as_val)).evalf()
sp.Eq(sp.symbols('N_{photons\\,per\\,pulse}'), N_photons_per_pulse)
```

```python
f_rep = 76e6  # repitition rate in Hz
produced_photon_flux = N_photons_per_pulse*f_rep
sp.Eq(sp.symbols('Produced\\,photon\\,flux'), sp.Float(produced_photon_flux,3))
```

```python
QE_factor = 0.5 # quantum efficiency of detector
loss_factor = 0.1 # losses in the system
int_time_num = 100e-3  # integration time in seconds
detected_counts = produced_photon_flux*QE_factor*loss_factor*int_time_num
sp.Eq(sp.symbols('Detected\\,Counts'), sp.Float(detected_counts,3))
```
