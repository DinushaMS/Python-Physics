```python
import sympy as sp
```

```python
x,y,z,t = sp.symbols('x y z t', real = True)
E10, E20, E30, Eas0 = sp.symbols('E_{10} E_{20} E_{30} E_{as0}')
Eas0 = sp.Function('E_{as0}')(z,t)
k1, k2, k3, kas, delk, om1, om2, om3, omas = sp.symbols('k_1 k_2 k_3 k_{as} \\Delta{k} \\omega_1 \\omega_2 \\omega_3 \\omega_{as}',real = True, nonzero=True)
eps0, chi3 = sp.symbols('\\epsilon_0 \\chi^{(3)}', real = True)
n1, n2, n3, nas, c, L = sp.symbols('n_1 n_2 n_3 n_{as} c L', real = True, nonzero=True, positive = True)
xi1,xi2,xi3,xias,tau,w0 = sp.symbols('xi_1 xi_2 xi_3 xi_as tau w_0', real=True, positive=True)
A1, A2, A3, Aas = sp.symbols('A_1 A_2 A_3 A_{as}', real=True, positive=True)
```

```python
E1 = sp.Rational(1,2)*E10*sp.exp(sp.I*(k1*z-om1*t))
E1 = E1 + sp.conjugate(E1)
E2 = sp.Rational(1,2)*E20*sp.exp(sp.I*(k2*z-om2*t))
E2 = E2 + sp.conjugate(E2)
E3 = sp.Rational(1,2)*E30*sp.exp(sp.I*(k3*z-om3*t))
E3 = E3 + sp.conjugate(E3)
Eas = sp.Rational(1,2)*Eas0*sp.exp(sp.I*(kas*z-omas*t))
Eas = Eas + sp.conjugate(Eas)
sp.Eq(sp.symbols('E_{as}(t)'),Eas)
```

```python
E_tot = E1+E2+E3
P3_tot = (eps0*chi3*E_tot**3).expand()
sel_E3_product = E10*sp.conjugate(E20)*E30
P3_as = P3_tot.subs(sel_E3_product,1).subs({E10:0, E20:0, E30:0})*sel_E3_product
P3_as = P3_as.subs(sp.exp(-sp.I*(om1-om2+om3)*t).expand(),sp.exp(-sp.I*omas*t))
sp.Eq(sp.symbols('P^{(3)}_{as}(t)'),P3_as)
```

```python
ddEasdz = sp.diff(Eas,z,2)
sp.Eq(sp.symbols('\\nabla^2E_{as}'),ddEasdz)
```

```python
ddEasdz_re = ddEasdz.subs({sp.conjugate(Eas0):0, sp.diff(sp.conjugate(Eas0),z,1):0, sp.diff(sp.conjugate(Eas0),z,2):0}).collect(sp.exp(sp.I*(omas*t-kas*z)))
ddEasdz_re = ddEasdz_re.subs(sp.diff(Eas0,z,2),0)
sp.Eq(sp.symbols('\\nabla^2E_{as}'),ddEasdz_re)
```

```python
ddEasdt = sp.diff(Eas,t,2)
sp.Eq(sp.symbols('\\frac{\\partial^2E_{as}}{\\partial{t^2}}'),ddEasdt)
```

```python
ddEasdt_re = ddEasdt.subs({sp.conjugate(Eas0):0, sp.diff(sp.conjugate(Eas0),t,1):0, sp.diff(sp.conjugate(Eas0),t,2):0}).collect(sp.exp(sp.I*(omas*t-kas*z)))
ddEasdt_re = ddEasdt_re.subs(sp.diff(Eas0,t,2),0)
sp.Eq(sp.symbols('\\frac{\\partial^2E_{as}}{\\partial{t^2}}'),ddEasdt_re)
```

```python
ddP3_asdt = sp.diff(P3_as,t,2).simplify()
sp.Eq(sp.symbols('\\frac{\\partial^2P^{(3)}_{as}}{\\partial{t^2}}'),ddP3_asdt)
```

```python
#wave equation
we_lhs = (ddEasdz_re-nas**2*ddEasdt_re/c**2)
we_lhs = (we_lhs*sp.exp(-sp.I*(kas*z-omas*t))*2).expand()
we_rhs = ddP3_asdt/eps0/c**2
we_lhs = we_lhs.subs(omas*nas/c,kas)
we_rhs = (we_rhs*sp.exp(-sp.I*(kas*z-omas*t))*2).expand().simplify()
we_rhs = we_rhs.subs(k1-k2+k3-kas,delk ,nonzero=True)
sp.Eq(we_lhs,we_rhs)
```

```python
#time evolution of Eas is smaller than spatial evolution
we_lhs = we_lhs.subs(sp.diff(Eas0,t,1),0)*(-sp.I/2/kas)
we_rhs = we_rhs*(-sp.I/2/kas)
sp.Eq(we_lhs,we_rhs)
```

```python
#integrate wrt z from 0 to L
we_lhs_int = sp.Integral(we_lhs,(z,0,L)).doit()
we_rhs_int = sp.Integral(we_rhs,(z,0,L)).doit().simplify()
sp.Eq(we_lhs_int,we_rhs_int)
```

```python
#let Eas(0,t) = 0
we_lhs_int = we_lhs_int.subs(Eas0.subs(z,0),0)
sp.Eq(we_lhs_int,we_rhs_int)
```

```python
#intensity
Ias = nas*eps0*c*we_rhs_int*sp.conjugate(we_rhs_int)/2
Ias = Ias.subs(E10*sp.conjugate(E10),sp.Abs(E10)**2)
Ias = Ias.subs(E20*sp.conjugate(E20),sp.Abs(E20)**2)
Ias = Ias.subs(E30*sp.conjugate(E30),sp.Abs(E30)**2)
sp.Eq(sp.symbols('I_{as}'),Ias)
```

```python
L = sp.symbols('L',real=True)
temp = sp.exp(sp.I*delk*L)-1
temp = (temp*sp.conjugate(temp)).expand()
temp = temp.subs(sp.exp(sp.I*delk*L),sp.cos(delk*L)+sp.I*sp.sin(delk*L)).simplify()
sp.Eq(sp.Abs(sp.exp(sp.I*delk*L)-1)**2,temp)
```

```python
Ias2 = Ias.subs(sp.exp(sp.I*delk*L),sp.cos(delk*L)+sp.I*sp.sin(delk*L)).simplify()
sp.Eq(sp.symbols('I_{as}'),Ias2)
```

```python
###################REWRITE ABOVE RESULT##########################
Ias3 = 9*chi3**2*eps0*omas**4*nas*sp.Abs(E10**2*E20**2*E30**2)*sp.sin(L*delk/2)**2/(8*delk**2*c**3*kas**2)
sp.Eq(sp.symbols('I_{as}'),Ias3)
```

```python
Ias4 = Ias3.subs(sp.sin(L*delk/2)**2,(L*delk/2)**2*sp.sinc(L*delk/2)**2)
sp.Eq(sp.symbols('I_{as}'),Ias4)
```

```python
Ias_max = Ias4.subs(sp.sinc(L*delk/2)**2,1)
sp.Eq(sp.symbols('I_{as\\,max}'),Ias_max)
```

```python
E10_sub = A1*sp.exp(-2*sp.log(2)*t**2/tau**2)*sp.exp(-(x**2+y**2)/w0**2)
E20_sub = A2*sp.exp(-2*sp.log(2)*t**2/tau**2)*sp.exp(-(x**2+y**2)/w0**2)
E30_sub = A3*sp.exp(-2*sp.log(2)*t**2/tau**2)*sp.exp(-(x**2+y**2)/w0**2)
sp.Eq(sp.symbols('E_{10}'), E10_sub)
```

```python
Ias_max = Ias_max.subs({E10:E10_sub,E20:E20_sub,E30:E30_sub})
sp.Eq(sp.symbols('I_{as\\,max}'),Ias_max)
```

```python
sp.Eq(xias, sp.Integral(sp.Integral(sp.Integral(sp.symbols('I_{as\\,max}'),(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo)))
```

```python
int1 = sp.Integral(sp.Integral(sp.Integral(sp.exp(-12*sp.log(2)*t**2/tau**2)*sp.exp(-6*(x**2+y**2)/w0**2),(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
sp.Eq(int1,int1.doit())
```

```python
xias_max = sp.Integral(sp.Integral(sp.Integral(Ias_max,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
sp.Eq(xias, xias_max)
```

```python
xias_max = xias_max.doit()
sp.Eq(xias, xias_max)
```

```python
I1 = sp.Rational(1/2)*n1*eps0*c*sp.Abs(E10)**2
I2 = sp.Rational(1/2)*n2*eps0*c*sp.Abs(E20)**2
I3 = sp.Rational(1/2)*n3*eps0*c*sp.Abs(E30)**2
sp.Eq(sp.symbols('I_1'), I1)
```

```python
I1 = sp.Rational(1/2)*n1*eps0*c*sp.Abs(E10_sub)**2
I2 = sp.Rational(1/2)*n2*eps0*c*sp.Abs(E20_sub)**2
I3 = sp.Rational(1/2)*n3*eps0*c*sp.Abs(E30_sub)**2
sp.Eq(sp.symbols('I_1'), I1)
```

```python
xi1_sub = sp.Integral(sp.Integral(sp.Integral(I1,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
xi2_sub = sp.Integral(sp.Integral(sp.Integral(I2,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
xi3_sub = sp.Integral(sp.Integral(sp.Integral(I3,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
sp.Eq(xi1, xi1_sub)
```

```python
xi1_sub = xi1_sub.doit()
xi2_sub = xi2_sub.doit()
xi3_sub = xi3_sub.doit()
sp.Eq(xi1, xi1_sub)
```

```python
xias_max_f = xias_max*(xi1/xi1_sub)*(xi2/xi2_sub)*(xi3/xi3_sub)
sp.Eq(xias, xias_max_f)
```

```python
xias_max_ff = xias_max_f.subs(kas,nas*omas/c)
sp.Eq(xias, xias_max_ff)
```

```python
xias_max_fff = xias_max_ff.subs(omas,2*sp.pi*c/sp.symbols('\\lambda_{as}'))
sp.Eq(xias, xias_max_fff)
```
