```python
import sympy as sp
```

```python
r,the,phi,t = sp.symbols('r theta phi t', real = True)
N, eps0, c, oms, omR, omL, q0, EL0, Es, ns = sp.symbols('N epsilon_0 c omega_s Omega_R omega_L q_0 E_{L0} E_s n_{s}', real = True, positive = True)
dadq = sp.symbols('\\left(\\frac{\\partial\\alpha}{\\partial{q}}\\right)_0', real = True)
Ps,ph2,ph3 = sp.symbols('P_s \\phi^{\'\'} \\phi^{\'\'\'}', real = True)
ks, kL = sp.symbols('k_{s} k_L', real = True)
Es0 = sp.Function('E_{s0}')(r,the,phi,t)
```

```python
def spherical_nabla_sq(f):
    return (1/r**2)*sp.diff(r**2*sp.diff(f,r),r) + (1/(r**2*sp.sin(the)))*sp.diff(sp.sin(the)*sp.diff(f,the),the) + (1/(r**2*sp.sin(the)**2))*sp.diff(f,phi,2)
```

```python
f = sp.Function('f')(r,the,phi,t)
spherical_nabla_sq(f)
```

```python
Ps_expr = sp.Rational(1,4)*N*dadq*q0*EL0*sp.exp(sp.I*(kL*r*sp.cos(the)-(omL-omR)*t)+sp.I*ph3).subs(omL-omR,oms)
sp.Eq(Ps,Ps_expr)
```

```python
Es_expr = sp.Rational(1,2)*Es0*sp.exp(sp.I*(ks*r-(omL-omR)*t)+sp.I*ph2).subs(omL-omR,oms)
sp.Eq(Es,Es_expr)
```

```python
nabla_sq_Es_expr = spherical_nabla_sq(Es_expr).simplify()
sp.Eq(sp.symbols(r'\nabla^{2}E_s'),nabla_sq_Es_expr)
```

```python
nabla_sq_Es_expr2 = nabla_sq_Es_expr.subs({sp.diff(Es0,r,2):0,sp.diff(Es0,the,2):0,sp.diff(Es0,phi,2):0,sp.diff(Es0,the,1):0}).simplify()
sp.Eq(sp.symbols(r'\nabla^{2}E_s'),nabla_sq_Es_expr2)
```

```python
ddEsdt_expr = sp.diff(Es_expr,t,2)
sp.Eq(sp.symbols(r'\frac{\partial^{2}{E_s}}{\partial{t^2}}'),ddEsdt_expr)
```

```python
ddPsdt_expr = sp.diff(Ps_expr,t,2)
sp.Eq(sp.symbols(r'\frac{\partial^{2}{P_s}}{\partial{t^2}}'),ddPsdt_expr)
```

```python
#wave equation
we_lhs = nabla_sq_Es_expr2 - (ns**2/c**2)*ddEsdt_expr
we_rhs = (1/(eps0*c**2))*ddPsdt_expr
we_lhs = (we_lhs*sp.exp(-sp.I*(ks*r-(omL-omR)*t)+sp.I*ph2)).expand()
sp.Eq(we_lhs,we_rhs)
```

```python
we_lhs = nabla_sq_Es_expr2.subs({sp.diff(Es0,r,1):0}) - (ns**2/c**2)*ddEsdt_expr.subs({sp.diff(Es0,t,2):0,sp.diff(Es0,t,1):0})
we_lhs = (we_lhs*sp.exp(-sp.I*(ks*r-oms*t)-sp.I*ph2)).expand()
we_lhs = we_lhs.subs(ks, ns*oms/c)*c*r/(sp.I*ns*oms)
we_rhs = (1/(eps0*c**2))*ddPsdt_expr
we_rhs = (we_rhs*sp.exp(-sp.I*(ks*r-oms*t)-sp.I*ph2)).expand()
we_rhs = we_rhs.subs(ks, ns*oms/c)*c*r/(sp.I*ns*oms)
sp.Eq(we_lhs,we_rhs)
```
