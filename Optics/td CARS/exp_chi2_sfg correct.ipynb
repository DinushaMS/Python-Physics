# Corrections

$$E_i=\frac{1}{2}E_{i0}e^{i(k_iz-\omega_it)}+c.c.$$

$$I_i=\frac{1}{2}n_i\epsilon_0c|E_{i0}|^2$$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y,z,t = sp.symbols('x y z t', real = True)
E10, E20, Esfg0 = sp.symbols('E_{10} E_{20} E_{sfg0}')
Esfg0 = sp.Function('E_{sfg0}')(z,t)
k1, k2, ksfg, delk, om1, om2, omsfg, wlsfg = sp.symbols('k_1 k_2 k_{sfg} \\Delta{k} \\omega_1 \\omega_2 \\omega_{sfg} \\lambda_{sfg}',real = True, nonzero=True)
eps0, chi2 = sp.symbols('\\epsilon_0 \\chi^{(2)}', real = True)
n1, n2, nsfg, c, L = sp.symbols('n_1 n_2 n_{sfg} c L', real = True, nonzero=True, positive = True)
xi1,xi2,xisfg,tau1,tau2,w0 = sp.symbols('xi_1 xi_2 xi_sfg tau_1 tau_2 w_0', real=True, positive=True)
A1, A2, Asfg = sp.symbols('A_1 A_2 A_{sfg}', real=True, positive=True)
I1, I2, Isfg = sp.symbols('I_1 I_2 I_{sfg}', real=True, positive=True)
E1, E2, Esfg = sp.symbols('E_1 E_2 E_{sfg}', real=True, positive=True)
```

```python
E1_expr = sp.Rational(1,2)*E10*sp.exp(sp.I*(k1*z-om1*t))
E1_expr = E1_expr + sp.conjugate(E1_expr)
E2_expr = sp.Rational(1,2)*E20*sp.exp(sp.I*(k2*z-om2*t))
E2_expr = E2_expr + sp.conjugate(E2_expr)
Esfg_expr = sp.Rational(1,2)*Esfg0*sp.exp(sp.I*(ksfg*z-omsfg*t))
Esfg_expr = Esfg_expr + sp.conjugate(Esfg_expr)
sp.Eq(Esfg,Esfg_expr)
```

```python
E_tot_expr = E1_expr+E2_expr
P2_tot_expr = (eps0*chi2*E_tot_expr**2).expand()
sel_E2_product_expr = E10*E20
P2_sfg_expr = P2_tot_expr.subs(sel_E2_product_expr,1).subs({E10:0, E20:0})*sel_E2_product_expr
P2_sfg_expr = P2_sfg_expr.subs(sp.exp(-sp.I*(om1+om2)*t).expand(),sp.exp(-sp.I*omsfg*t))
sp.Eq(sp.symbols('P^{(2)}_{sfg}(t)'),P2_sfg_expr)
```

```python
ddEsfgdz_expr = sp.diff(Esfg_expr,z,2)
sp.Eq(sp.symbols('\\nabla^2E_{sfg}'),ddEsfgdz_expr)
```

```python
ddEsfgdz_re_expr = ddEsfgdz_expr.subs({sp.conjugate(Esfg0):0, sp.diff(sp.conjugate(Esfg0),z,1):0, sp.diff(sp.conjugate(Esfg0),z,2):0}).collect(sp.exp(sp.I*(omsfg*t-ksfg*z)))
ddEsfgdz_re_expr = ddEsfgdz_re_expr.subs(sp.diff(Esfg0,z,2),0)
sp.Eq(sp.symbols('\\nabla^2E_{sfg}'),ddEsfgdz_re_expr)
```

```python
ddEsfgdt_expr = sp.diff(Esfg_expr,t,2)
sp.Eq(sp.symbols('\\frac{\\partial^2E_{sfg}}{\\partial{t^2}}'),ddEsfgdt_expr)
```

```python
ddEsfgdt_re_expr = ddEsfgdt_expr.subs({sp.conjugate(Esfg0):0, sp.diff(sp.conjugate(Esfg0),t,1):0, sp.diff(sp.conjugate(Esfg0),t,2):0}).collect(sp.exp(sp.I*(omsfg*t-ksfg*z)))
ddEsfgdt_re_expr = ddEsfgdt_re_expr.subs(sp.diff(Esfg0,t,2),0)
sp.Eq(sp.symbols('\\frac{\\partial^2E_{sfg}}{\\partial{t^2}}'),ddEsfgdt_re_expr)
```

```python
ddP2_sfgdt_expr = sp.diff(P2_sfg_expr,t,2).simplify()
sp.Eq(sp.symbols('\\frac{\\partial^2P^{(2)}_{sfg}}{\\partial{t^2}}'),ddP2_sfgdt_expr)
```

```python
#wave equation
we_lhs_expr = (ddEsfgdz_re_expr-nsfg**2*ddEsfgdt_re_expr/c**2)
we_rhs_expr = ddP2_sfgdt_expr/eps0/c**2
sp.Eq(we_lhs_expr,we_rhs_expr)
```

```python
we_lhs_expr = (we_lhs_expr*sp.exp(-sp.I*(ksfg*z-omsfg*t))*2).expand()
we_rhs_expr = (we_rhs_expr*sp.exp(-sp.I*(ksfg*z-omsfg*t))*2).expand().simplify()
we_rhs_expr = we_rhs_expr.subs(k1+k2-ksfg,delk ,nonzero=True)
sp.Eq(we_lhs_expr,we_rhs_expr)
```

```python
#time evolution of Eas is smaller than spatial evolution
we_lhs_expr = we_lhs_expr.subs(sp.diff(Esfg0,t,1),0)*(-sp.I/2/ksfg)
we_rhs_expr = we_rhs_expr*(-sp.I/2/ksfg)
sp.Eq(we_lhs_expr,we_rhs_expr)
```

```python
#integrate wrt z from 0 to L
we_lhs_int_expr = sp.Integral(we_lhs_expr,(z,0,L)).doit()
we_rhs_int_expr = sp.Integral(we_rhs_expr,(z,0,L)).doit().simplify()
sp.Eq(we_lhs_int_expr,we_rhs_int_expr)
```

```python
#let Eas(0,t) = 0
we_lhs_int_expr = we_lhs_int_expr.subs(Esfg0.subs(z,0),0)
sp.Eq(we_lhs_int_expr,we_rhs_int_expr)
```

```python
#intensity############################################################
Isfg_expr = sp.Rational(1,2)*nsfg*eps0*c*we_rhs_int_expr*sp.conjugate(we_rhs_int_expr)
#intensity############################################################
Ias_expr = Isfg_expr.subs(E10*sp.conjugate(E10),sp.Abs(E10)**2)
Ias_expr = Isfg_expr.subs(E20*sp.conjugate(E20),sp.Abs(E20)**2)
sp.Eq(sp.symbols('I_{sfg}'),Isfg_expr)
```

```python
temp = sp.exp(sp.I*delk*L)-1
temp = (temp*sp.conjugate(temp)).expand()
temp = temp.subs(sp.exp(sp.I*delk*L),sp.cos(delk*L)+sp.I*sp.sin(delk*L)).simplify()
sp.Eq(sp.Abs(sp.exp(sp.I*delk*L)-1)**2,temp)
```

```python
Isfg_expr2 = Isfg_expr.subs(sp.exp(sp.I*delk*L),sp.cos(delk*L)+sp.I*sp.sin(delk*L)).simplify()
sp.Eq(Isfg,Isfg_expr2)
```

```python
###################REWRITE ABOVE RESULT##########################
Isfg_expr3 = chi2**2*eps0*omsfg**4*nsfg*sp.Abs(E10**2*E20**2)*sp.sin(L*delk/2)**2/(2*delk**2*c**3*ksfg**2)
sp.Eq(sp.symbols('I_{sfg}'),Isfg_expr3)
```

```python
Isfg_in_E_expr = Isfg_expr3.subs(sp.sin(L*delk/2)**2,(L*delk/2)**2*sp.sinc(L*delk/2)**2)
sp.Eq(sp.symbols('I_{sfg}'),Isfg_in_E_expr)
```

```python
E10_expr = A1*sp.exp(-2*sp.log(2)*t**2/tau1**2)*sp.exp(-(x**2+y**2)/w0**2)
E20_expr = A2*sp.exp(-2*sp.log(2)*t**2/tau2**2)*sp.exp(-(x**2+y**2)/w0**2)
sp.Eq(E10, E10_expr)
```

```python
Isfg_in_A_expr = Isfg_in_E_expr.subs({E10:E10_expr,E20:E20_expr})
sp.Eq(sp.symbols('I_{sfg}'),Isfg_in_A_expr)
```

```python
sp.Eq(xisfg, sp.Integral(sp.Integral(sp.Integral(Isfg,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo)))
```

```python
int1 = sp.Integral(sp.Integral(sp.Integral(sp.exp(-4*sp.log(2)*t**2/tau1**2)*sp.exp(-4*sp.log(2)*t**2/tau2**2)*sp.exp(-4*(x**2+y**2)/w0**2),(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
sp.Eq(int1,int1.doit())
```

```python
xisfg_expr = sp.Integral(sp.Integral(sp.Integral(Isfg_in_A_expr,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
sp.Eq(xisfg, xisfg_expr)
```

```python
xisfg_expr = xisfg_expr.doit()
sp.Eq(xisfg, xisfg_expr)
```

```python
#intensity############################################################
I1_in_E_expr = sp.Rational(1,2)*n1*eps0*c*sp.Abs(E10)**2
I2_in_E_expr = sp.Rational(1,2)*n2*eps0*c*sp.Abs(E20)**2
#intensity############################################################
sp.Eq(I1, I1_in_E_expr)
```

```python
E10_abs_sq_in_I_expr = I1/I1_in_E_expr*sp.Abs(E10)**2
E20_abs_sq_in_I_expr = I2/I2_in_E_expr*sp.Abs(E20)**2
sp.Eq(sp.Abs(E10)**2, E10_abs_sq_in_I_expr)
```

```python
Ias_in_I_expr = sp.simplify(Isfg_in_E_expr/sp.Abs(E10)**2/sp.Abs(E20)**2)*E10_abs_sq_in_I_expr*E20_abs_sq_in_I_expr
sp.Eq(Isfg,Ias_in_I_expr)
```

```python
I1_in_A_expr = I1_in_E_expr.subs(E10,E10_expr)
I2_in_A_expr = I2_in_E_expr.subs(E20,E20_expr)
sp.Eq(I1, I1_in_A_expr)
```

```python
xi1_expr = sp.Integral(sp.Integral(sp.Integral(I1_in_A_expr,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
xi2_expr = sp.Integral(sp.Integral(sp.Integral(I2_in_A_expr,(t,-sp.oo,sp.oo)),(x,-sp.oo,sp.oo)),(y,-sp.oo,sp.oo))
sp.Eq(xi1, xi1_expr)
```

```python
xi1_expr = xi1_expr.doit()
xi2_expr = xi2_expr.doit()
sp.Eq(xi1, xi1_expr)
```

```python
xias_expr_f = xisfg_expr*(xi1/xi1_expr)*(xi2/xi2_expr)
sp.Eq(xisfg, xias_expr_f)
```

```python
xias_expr_ff = xias_expr_f.subs(ksfg,nsfg*omsfg/c)
sp.Eq(xisfg, xias_expr_ff)
```

```python
xisfg_expr_fff = xias_expr_ff.subs(omsfg,2*sp.pi*c/wlsfg)
sp.Eq(xisfg, xisfg_expr_fff)
```

```python
P1_val, P2_val = 100e-3, 300e-3 # power in W (100 mW)
tau1_val, tau2_val = 100e-15, 200e-15  # seconds
f_rep = 76e6  # repetition rate in Hz (76 MHz)
xi1_val, xi2_val = P1_val/f_rep, P2_val/f_rep
sp.Eq(xi1, xi1_val)
```

```python
w0_val = (200e-9)/2  # meters
chi2_val = 1e-12  # m/V
L_val = 200e-9  # meters
wlsfg_val = 450e-9  # meters
n1_val, n2_val, nsfg_val = 2.5, 2.5, 2.5  # refractive indices
delk_val = 0  # phase mismatch
eps0_val = 8.854187817e-12  # F/m
c_val = 3e8  # m/s

xisfg_num = xisfg_expr_fff.evalf(subs={tau1:tau1_val, tau2:tau2_val, w0:w0_val, chi2:chi2_val, L:L_val, wlsfg:wlsfg_val, n1:n1_val, n2:n2_val, nsfg:nsfg_val, xi1:xi1_val, xi2:xi2_val, delk:delk_val, eps0:eps0_val, c:c_val})
sp.Eq(xisfg, xisfg_num)
```

```python
h_bar = 1.0545718e-34  # J·s
omega_sfg_val = 2*sp.pi*c_val/wlsfg_val  # rad/s
N_photons_per_pulse = (xisfg_num/(h_bar*omega_sfg_val)).evalf()
sp.Eq(sp.symbols('N_{photons\\,per\\,pulse}'), int(sp.N(N_photons_per_pulse).round(0)))
```

```python
f_rep = 76e6  # repitition rate in Hz
produced_photon_flux = N_photons_per_pulse*f_rep
sp.Eq(sp.symbols('Produced\\,photon\\,flux'), sp.Float(produced_photon_flux,3))
```

```python
QE_factor = 0.5 # quantum efficiency of detector
loss_factor = 0.1 # losses in the system
int_time_num = 100e-3  # integration time in seconds
detected_counts = produced_photon_flux*QE_factor*(1-loss_factor)*int_time_num
sp.Eq(sp.symbols('Detected\\,Counts'), sp.Float(detected_counts,3))
```
