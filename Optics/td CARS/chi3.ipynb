```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['ieee'])
plt.rcParams.update({'font.size': 20})
```

```python
c = 3e8 # Speed of light in m/s
nu_arr = np.linspace(0, 1000, 1000) # Frequency array from 10-1000 1/cm
om_arr = 2 * np.pi * c * nu_arr * 100 # Convert to angular frequency in rad/s
```

```python
OmR = 2 * np.pi * c * 500 * 100 # Resonance frequency at 300 1/cm
T2 = 500e-15 # Dephasing time in seconds
chi3 = 1/(OmR**2-om_arr**2-2j*om_arr/T2) # Example chi3 function
plt.figure(figsize=(8,5))
plt.plot(nu_arr, np.real(chi3), label=r'$\Re[\chi^{(3)}]$')
plt.plot(nu_arr, np.imag(chi3), label=r'$\Im[\chi^{(3)}]$')
plt.plot(nu_arr, np.abs(chi3), label=r'$|\chi^{(3)}|$')
plt.legend()
plt.xlabel(r'$\omega$'+' [1/cm]')
plt.ylabel(r'$\chi^{(3)}$' + '[a.u.]')
plt.title('Third-Order Susceptibility vs Frequency')
plt.show()
```
