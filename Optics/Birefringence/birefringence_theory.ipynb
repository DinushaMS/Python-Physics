```python
import sys
sys.path.append("..")
from functions import bbo_n, gen_n_circle, gen_k_array, gen_del_k_plot, n_squared
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta_array = np.linspace(0*np.pi/180,90*np.pi/180,2000)
wl_array = [1.064,0.532]

fig,ax = gen_n_circle(theta_array, wl_array)
ax.text(1.8,-0.2,'$x$')
ax.text(0.1,1.8,'$z$')
ax.legend(loc='upper right')
ax.set_title('Refractive Index Surface of BBO Crystal')
```

```python
kk_array = gen_k_array(theta_array,wl_array)
plt.figure(figsize=(7, 4),dpi=300)
for i,wl in enumerate(wl_array):
    plt.plot(theta_array*180/np.pi,kk_array[:,i,0]/1e5,label = '$k_o$'+f'({wl})')
    plt.plot(theta_array*180/np.pi,kk_array[:,i,1]/1e5,label = '$k_e$'+f'({wl})')
plt.legend()
plt.ylabel('$k$'+'('+'$\\times$'+'$10^5$ rad/m)')
plt.xlabel('$\\theta$(deg.)')
plt.plot()
```
