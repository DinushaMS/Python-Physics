```python
import numpy as np
from scipy.optimize import minimize

# Example function
def func(a, b, c, d, e):
    # Replace this with your actual function
    return a + 2*b - c + d*e

# Expected value
y_act = 10.0

# Define RMS loss function
def rms_loss(params):
    a, b, c, d, e = params
    y_exp = func(a, b, c, d, e)
    rms = np.sqrt((y_exp - y_act)**2)  # For single value, sqrt((y_exp - y_act)^2) = |y_exp - y_act|
    return rms
```

```python
# Initial guess for parameters
initial_guess = [1, 1, 1, 1, 1]

# Run optimization
result = minimize(rms_loss, initial_guess, method='Nelder-Mead')

# Optimized parameters
optimized_params = result.x
optimized_value = func(*optimized_params)

print("Optimized parameters:", optimized_params)
print("Function value with optimized parameters:", optimized_value)
print("RMS error:", rms_loss(optimized_params))
```
