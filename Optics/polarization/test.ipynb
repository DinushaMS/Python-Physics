```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
from functions import bbo_n
```

```python
t = np.linspace(0, 2*np.pi, 500)

Ex = np.sin(t)
Ey = np.sin(t + np.pi/4)  # Circular polarization with zero amplitude in Ey

plt.figure()
plt.plot(Ex, Ey)
plt.gca().set_aspect('equal')
plt.xlabel("Ex")
plt.ylabel("Ey")
plt.title("Circular Polarization")
plt.show()
```
