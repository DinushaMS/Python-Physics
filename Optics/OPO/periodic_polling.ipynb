```python
import sys
sys.path.append("..")
from functions import ln_n
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','ieee'])
plt.rcParams.update({'font.size': 12})
```

```python
#For OPA in LN
wl_p = 0.8 #um
wl_s = np.linspace(0.95,1.55,100) #um
wl_i = 1/(1/wl_p - 1/wl_s) #um
n_s = ln_n(wl_s)
n_i = ln_n(wl_i)
n_p = ln_n(wl_p)
k_s = 2*np.pi*n_s/wl_s
k_i = 2*np.pi*n_i/wl_i
k_p = 2*np.pi*n_p/wl_p
del_k = k_p - k_s - k_i
lc = np.pi/del_k #um
```

```python
plt.plot(lc,wl_s*1e3,label="Signal")
#plt.plot(lc,wl_i*1e3,label="Idler")
plt.legend()
plt.ylabel("Wavelength (nm)")
plt.xlabel("Coherence Length (um)")
plt.title(r"OPA in LN ($\lambda_p$="+f"{wl_p}"+r"$\mu$m, $21^o$C)",fontsize=12)
plt.show()
```
