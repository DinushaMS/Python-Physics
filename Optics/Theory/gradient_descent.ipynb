```python
import numpy as np
```

```python
def gradient_descent(f, grad_f, x0, lr=0.1, n_iter=100):
    x = np.array(x0, dtype=float)
    history = []

    for i in range(n_iter):
        grad = grad_f(x)
        x -= lr * grad
        history.append(x.copy())

    return x, np.array(history)
```

```python
# Function and gradient
f = lambda x: np.dot(x-2, x-2)
grad_f = lambda x: 2 * x - 4

x_min, path = gradient_descent(f, grad_f, x0=[3, -4], lr=0.1)

print("Minimum at x =", x_min)
```
