#Sellmeier Equation
The Sellmeier equation is an empirical formula that relates the refractive index of a material to the wavelength of light, used to model the dispersion of light. The most common form of the equation is a sum of terms, each representing an absorption resonance, and uses experimentally determined coefficients ($B_i$ and $C_i$) that are specific to the material:
$$n^2(\lambda)=1+\sum\left[\frac{B_i\lambda^2}{\lambda^2-C_i}\right]$$
| Material | B | C |
| -------- | -------- | -------- |
| $BBO_o$ | $$1.7359\lambda ^{-2}\\0.01878\lambda^{-2}\\-0.01354$$ |$$\lambda^2-1\\0.01822\\\lambda^2-1$$ |
| $BBO_e$ | $$1.3753\lambda ^{-2}\\0.01224\lambda^{-2}\\-0.01516$$ |$$\lambda^2-1\\0.01667\\\lambda^2-1$$ |

```python
import sys
sys.path.append("..")
from functions import n_squared
import numpy as np
import matplotlib.pyplot as plt
```

```python
wl_array = np.linspace(0.9,1.1,1000)
n_o_array = np.array([])
n_e_array = np.array([])
for wl in wl_array:
    B_o = np.array([1.7359/wl**2,0.01878/wl**2,-0.01354])
    C_o = np.array([wl**2-1,0.01822,wl**2-1])
    B_e = np.array([1.3753/wl**2,0.01224/wl**2,-0.01516])
    C_e = np.array([wl**2-1,0.01667,wl**2-1])
    n_o_array = np.append(n_o_array,(n_squared(wl,B_o,C_o)))
    n_e_array = np.append(n_e_array,(n_squared(wl,B_e,C_e)))
plt.plot(wl_array,n_o_array,label="n_o")
plt.plot(wl_array,n_e_array,label="n_e")
plt.legend()
```
